```python
# 1. Correlation matrix of symptom duration and severity
correlation_matrix = data1[['Age', 'Duration of Symptoms (months)', 'Y-BOCS Score (Obsessions)', 'Y-BOCS Score (Compulsions)']].corr()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matric of OCD severity and Duration of symptoms')
plt.show()
```

```python
# 2. Correlation between Symptom duration with anxiety and depression levels
data1['Anxiety Diagnosis'] = data1['Anxiety Diagnosis'].map({'Yes': 1, 'No': 0})
data1['Depression Diagnosis'] = data1['Depression Diagnosis'].map({'Yes': 1, 'No': 0})
correlation_matrix = data1[['Anxiety Diagnosis', 'Depression Diagnosis', 'Duration of Symptoms (months)']].corr()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation between Symptom duration with anxiety and depression levels')
plt.show()
```

```python
# 2. Demographic factors and analyze severity
# Gender analysis
gender_analysis = data1.groupby('Gender')[['Y-BOCS Score (Obsessions)',
                                         'Y-BOCS Score (Compulsions)',
                                         'Duration of Symptoms (months)']].agg(['mean', 'std'])

# Ethnicity analysis
ethnicity_analysis = data1.groupby('Ethnicity')[['Y-BOCS Score (Obsessions)',
                                               'Y-BOCS Score (Compulsions)',
                                               'Duration of Symptoms (months)']].agg(['mean', 'std'])

# Visualize gender differences
plt.figure(figsize=(6, 6))
sns.boxplot(x='Gender', y='Y-BOCS Score (Obsessions)', data=data1)
plt.title('OCD Obsession Severity by Gender')
plt.show()
```

```python
# 3. Analyze treatment outcomes
treatment_analysis = data1.groupby('Medications')[['Y-BOCS Score (Obsessions)',
                                                  'Y-BOCS Score (Compulsions)']].agg(['mean', 'std'])

# Visualize treatment outcomes
plt.figure(figsize=(6,6))
sns.boxplot(data=data1, x='Medications', y='Y-BOCS Score (Obsessions)')
plt.title('OCD Severity by Medication Type')
plt.show()
```

```python
data1.drop_duplicates(inplace=True)
```

```python
Y_bocs=data1[['Age','Obsession Type','Compulsion Type','Y-BOCS Score (Obsessions)','Y-BOCS Score (Compulsions)']]
```

```python
Y_bocs.head()
```

```python
def categorize_ocd(score):
    if 0 <= score <= 7:
        return "Subclinical OCD"
    elif 8 <= score <= 15:
        return "Mild OCD"
    elif 16 <= score <= 23:
        return "Moderate OCD"
    elif 24 <= score <= 31:
        return "Severe OCD"
    elif 32 <= score <= 40:
        return "Extreme OCD"
    else:
        return "Unknown"

Y_bocs['OCD Severity (Obsessions)'] = Y_bocs['Y-BOCS Score (Obsessions)'].apply(categorize_ocd)

Y_bocs['OCD Severity (Compulsions)'] = Y_bocs['Y-BOCS Score (Compulsions)'].apply(categorize_ocd)

Y_bocs[['Age', 'Obsession Type', 'Compulsion Type', 'Y-BOCS Score (Obsessions)', 'OCD Severity (Obsessions)', 'Y-BOCS Score (Compulsions)', 'OCD Severity (Compulsions)']]
```

```python
Y_bocs.to_csv ('OCD Y-Bocs scores categories')
```
